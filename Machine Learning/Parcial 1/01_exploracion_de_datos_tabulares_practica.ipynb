Imagina que estamos interesados en predecir las especies de pingüinos basándonos en dos de sus medidas corporales: la longitud y profundidad del pico. Primero queremos hacer una exploración de los datos para tener una idea de ellos.

Los datos se encuentran en `penguins.csv`, cárgalos con `pandas` en un `DataFrame`.

```python
import pandas as pd

df = pd.read_csv("penguins.csv")
```

¿Cuáles son las diferentes especies de pingüinos disponibles en el conjunto de datos y cuántas muestras de cada especie hay? Sugerencia: selecciona la columna correcta y utiliza el método [`value_counts`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.value_counts.html).

```python
# Hay 3 especies diferentes. Hay 152 de Adelie, 124 de Gentoo y 68 de Chinstrap
df["species"].value_counts()
```

Dibuja los histogramas para las características numéricas.

```python
_ = df.hist(figsize=(10, 10))
```

Muestra la distribución de las características para cada clase. Sugerencia: utiliza [`seaborn.pairplot`](https://seaborn.pydata.org/generated/seaborn.pairplot.html).

```python
import seaborn as sns

n_samples = 300
columns = ["bill_length_mm",	"bill_depth_mm",	"flipper_length_mm",	"body_mass_g"]
_ = sns.pairplot(data=df[:n_samples],vars=columns,hue="species",plot_kws={"alpha": 0.2},diag_kind="hist")
```

Al observar estas distribuciones, ¿qué tan difícil crees que sería clasificar a los pingüinos solo utilizando `"bill_length_mm"` y `"bill_depth_mm"`?

```python
# No muy difícil, los conjuntos están definidos
columns = ["bill_length_mm",	"bill_depth_mm"]
_ = sns.pairplot(data=df[:n_samples],vars=columns,hue="species",plot_kws={"alpha": 0.2},diag_kind="hist")
```
