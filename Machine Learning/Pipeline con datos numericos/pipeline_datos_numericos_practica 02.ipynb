El objetivo de esta pr치ctica es comparar el rendimiento del clasificador `LogisticRegression` implementado en clase (aproximadamente 80% de precisi칩n) con algunos clasificadores base simples. El clasificador base m치s simple es uno que siempre predice la misma clase, independientemente de los datos de entrada.

- 쮺u치l ser칤a el puntaje de un modelo que siempre predice `' >50K'`?
- 쮺u치l ser칤a el puntaje de un modelo que siempre predice `' <=50K'`?
- 쯋n accuracy del 80% es un buen puntaje para este problema?

Utiliza un `DummyClassifier` y realiza una divisi칩n de los datos en entrenamiento y prueba para evaluar su precisi칩n (accuracy) en el conjunto de prueba. Este [enlace](https://scikit-learn.org/stable/modules/model_evaluation.html#dummy-estimators) muestra algunos ejemplos de c칩mo evaluar el rendimiento de generalizaci칩n de estos modelos base.

```python
import pandas as pd

adult_census = pd.read_csv("adult.csv")
```

Primero separamos el dataset en 'data', y 'target'.

```python
target_name = "income"
target = adult_census[target_name]
data = adult_census.drop(columns=target_name)
```

Seleccionamos s칩lo las caracter칤sticas num칠ricas.

```python
numerical_columns = ["age", "capital.gain", "capital.loss", "hours.per.week"]

data_numeric = data[numerical_columns]
```

Separamos los datos y el objetivo en entrenamiento 'train' y prueba 'test'.

```python
from sklearn.model_selection import train_test_split

train_data, test_data, train_target, test_target = train_test_split(
    data_numeric, target, random_state=42, test_size=0.25
)
```

Usa el modelo `DummyClassifier` de tal forma que siempre clasifique `' >50K'`. Calcula el accuracy utilizando los datos de prueba. Repite el proceso pero ahora usa el modelo `DummyClassifier` de tal forma que siempre clasifique `' <=50K'`.
Nota: puedes configurar el par치metro `strategy`de `DummyClassifier` para lograr el comportamiento deseado del clasificador.

```python
from sklearn.dummy import DummyClassifier

strategies = ['stratified', 'uniform', 'most_frequent']

for strategy in strategies:
    dummy = DummyClassifier(strategy=strategy, random_state=42)
    dummy.fit(train_data, train_target)
    accuracy = dummy.score(test_data, test_target)
    print(f"Accuracy con estrategia '{strategy}': {accuracy:.4f}")
```
