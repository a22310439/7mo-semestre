El objetivo de esta pr치ctica es ajustar un modelo similar al visto en clase para familiarizarse con la manipulaci칩n de objetos de scikit-learn y, en particular, con la API .fit/.predict/.score.

Para empezar, cargamos el set de datos:

```python
import pandas as pd

adult_census = pd.read_csv("adult.csv")
data = adult_census.drop(columns="income")
target = adult_census["income"]

data = adult_census.drop(columns=["workclass", "fnlwgt", "education", "marital.status", "occupation", "relationship", "race", "sex", "native.country", "income"])
data.head()
```

Anteriormente usamos `model = KNeighborsClassifier()`. Todos los modelos de scikit-learn pueden crearse sin argumentos. Esto es conveniente porque significa que no necesitas entender todos los detalles de un modelo antes de empezar a usarlo.

Uno de los par치metros de `KNeighborsClassifier` es `n_neighbors`. Este controla el n칰mero de vecinos que vamos a usar para hacer una predicci칩n para un nuevo dato.

쮺u치l es el valor predeterminado del par치metro `n_neighbors`?

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()
print(model.n_neighbors)
```

Crea un modelo `KNeighborsClassifier` con `n_neighbors=50`

```python
model = KNeighborsClassifier(50)
print(model.n_neighbors)
```

Entrena el modelo con los datos y el objetivo definidos anteriormente.

```python
_ = model.fit(data, target)
```

Usa el modelo para hacer predicciones sobre los datos y compara si las predicciones son iguales al objetivo. Muestra los primeros 10 resultados. Pista: se debe mostrar 10 veces 'True' o 'False' dependiendo de si la predicci칩n coincide o no con el objetivo.

```python
predictions = model.predict(data)
predictions[:10]
```

```python
target[:10] == predictions[:10]
```

Calcula el 'accuracy' del modelo utilizando la funci칩n 'score'.

```python
accuracy = model.score(data, target)
accuracy
```
