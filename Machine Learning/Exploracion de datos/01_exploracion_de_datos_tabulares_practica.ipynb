Sugerencia: selecciona la columna correcta y utiliza el m√©todo [`value_counts`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.value_counts.html).

Dibuja los histogramas para las caracter√≠sticas num√©ricas.

Muestra la distribuci√≥n de las caracter√≠sticas para cada clase. Sugerencia: utiliza [`seaborn.pairplot`](https://seaborn.pydata.org/generated/seaborn.pairplot.html).

```python
import seaborn as sns

n_samples = 300
columns = ["bill_length_mm",	"bill_depth_mm",	"flipper_length_mm",	"body_mass_g"]
_ = sns.pairplot(data=df[:n_samples],vars=columns,hue="species",plot_kws={"alpha": 0.2},diag_kind="hist")
```

Al observar estas distribuciones, ¬øqu√© tan dif√≠cil crees que ser√≠a clasificar a los ping√ºinos solo utilizando `"bill_length_mm"` y `"bill_depth_mm"`?

```python
# No muy dif√≠cil, los conjuntos est√°n definidos
columns = ["bill_length_mm",	"bill_depth_mm"]
_ = sns.pairplot(data=df[:n_samples],vars=columns,hue="species",plot_kws={"alpha": 0.2},diag_kind="hist")
```
