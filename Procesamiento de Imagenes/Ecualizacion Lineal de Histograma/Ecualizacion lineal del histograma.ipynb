```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
img = cv.imread('corridor.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv2_imshow(img)
```

```python
h = plt.hist(img.flatten(), 256, range=[0,256])
```

```python
H = plt.hist(img.flatten(), 256, range=[0,256], cumulative=True)
```

```python
# Primero calcular filas y columnas
M,N = img.shape

#Matriz de ceros de M,N
imgEq = np.zeros([M,N])

H_ = H[0]

for x in range(M):
    for y in range(N):
      imgEq[x,y] = (H_[img[x,y]]*255)/(M*N)

#Mostrar imagen
plt.imshow(imgEq, cmap='gray')
```

```python
plt.hist(imgEq.flatten(), 256, range=[0,256])
```

```python
plt.hist(imgEq.flatten(), 256, range=[0,256], cumulative=True)
```
