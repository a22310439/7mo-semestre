```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

img = cv.imread('lennacolor.png')
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv2_imshow(img)
cv2_imshow(img_gray)
```

```python
# Filtro Box
img_blur = cv.blur(img, ksize=(9,9))
cv2_imshow(img_blur)
```

```python
# Filtro Gausseano
img_gauss = cv.GaussianBlur(img, ksize=(5,5), sigmaX=0)
cv2_imshow(img_gauss)
```

```python
# Filtro de Diferencia (Mexican Hat)
kernel = np.array([[0, 0, -1, 0, 0],
                   [0, -1, -2, -1, 0],
                   [-1, -2, 16, -2, -1],
                   [0, -1, -2, -1, 0],
                   [0, 0, -1, 0, 0]])
img_hat = cv.filter2D(img, ddepth=-1, kernel=kernel)
cv2_imshow(img_hat)
```

```python
# Filtro de diferencia a filtro Gausseano
img_diff_gauss = cv.filter2D(img_gauss, ddepth=-1, kernel=kernel)
cv2_imshow(img_diff_gauss)
```

```python
# Escala de grises con filtro de diferencia 5x5
img_gray_diff = cv.filter2D(img_gray, ddepth=-1, kernel=kernel)
cv2_imshow(img_gray_diff)
```

```python
# Escala de grises con filtro Gausseano y luego diferencia 5x5
img_gray_gauss = cv.GaussianBlur(img_gray, ksize=(5,5), sigmaX=0)
img_gray_gauss_diff = cv.filter2D(img_gray_gauss, ddepth=-1, kernel=kernel)
cv2_imshow(img_gray_gauss_diff)
```
