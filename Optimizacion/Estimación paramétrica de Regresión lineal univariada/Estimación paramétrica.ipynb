```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Cargar datos
data = pd.read_csv('dataset_RegresionLineal.csv')
data.head()
```

```python
x = np.array(data['x'])
y = np.array(data['y'])
m = np.size(x)

#Graficar datos
def graficarDatos():
  plt.plot(x, y, 'o', color='w', mec='black')
  plt.title('Grafica de dispersion')
  plt.xlabel('x: area m^2')
  plt.ylabel('y: precio $')
graficarDatos()
plt.show()
```

```python
#ESTIMACION PARAMETRICA
x_media = np.mean(x)
y_media = np.mean(y)

# Calcular β₁ (a1) usando: β₁ = cov(x,y) / var(x)
covarianza_xy = np.mean((x - x_media) * (y - y_media))
varianza_x = np.mean((x - x_media) ** 2)
a1 = covarianza_xy / varianza_x

# Calcular β₀ (a0) usando: β₀ = ȳ - β₁·x̄
a0 = y_media - a1 * x_media

# Calcular hipótesis final
h = a0 + a1 * x

# Calcular error (costo)
J = (1 / (2 * m)) * np.sum(np.square(h - y))
```

```python
#Graficar datos finales o modelo final
graficarDatos()
plt.plot(x, h, color='r', label=f'y = {a0:.2f} + {a1:.2f}x')
plt.legend()
plt.show()
```

```python
# Resultados
print('Error final: ', J)
print('a0: ', a0)
```
