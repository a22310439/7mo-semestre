```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Cargar datos
data = pd.read_csv('dataset_RegresionLinealMultivariable.csv')
data.head()
```

```python
# Separar datos en X y Y
x = data.drop(columns=['y'])
y = data['y'].values
```

```python
model = LinearRegression()
model.fit(x, y)

print("b0=", model.intercept_,
      "b1=", model.coef_[0],
      "b2=", model.coef_[1])
```

```python
# Solucion con estimacion parametrica
x_ones = np.hstack([np.ones((x.shape[0], 1)), x])
b = np.linalg.inv(x_ones.T @ x_ones) @ x_ones.T @ y
print("b0=", b[0],
      "b1=", b[1],
      "b2=", b[2])
```
